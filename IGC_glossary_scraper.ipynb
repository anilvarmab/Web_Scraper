```python
from bs4 import BeautifulSoup
import requests
import urllib.request
import re
```

```python
url = "https://www.healthcare.gov/glossary/"
```

```python
page = urllib.request.urlopen(url) # conntect to website
```

```python
#Parse the HTML page and check the HTML tags and content 
soup = BeautifulSoup(page, 'html.parser')
print(soup.select)
```

```python
glossary=soup.select('div > li')
r=glossary[6]
print(r)
```

```python
#Terms are available as description in the a tag s
#making a of List Terms from a tag descriptions
term_list=list()
for li in soup.find_all(class_="col-sm-4 gpost"):
    x=li.a.text
    term_list.append(x)
print(term_list)
```

```python
#Making a list of href tags for hyperlink extensions  
href_list=list()
for li in soup.find_all(class_="col-sm-4 gpost"):
    x=li.a.get('href')
    href_list.append(x)
print(href_list)
```

```python
#Checking content from one hyperlink to understand tags
url1 = "https://www.healthcare.gov/glossary/abortion-services/"
page1=urllib.request.urlopen(url1)
soup1=BeautifulSoup(page1, 'html.parser')
print(soup1.select)
```

```python
#Term description available under P tag
href1_list=list()
for div in soup1.find_all(class_="col-sm-10"):
    y=div.find('p')
    href1_list.append(y)
print(href1_list)
```

```python
#Function to read each term hyperlink and extract the term description from p tag
term_desc=list()
for x in href_list:
    url2 = "https://www.healthcare.gov" + x
    page1=urllib.request.urlopen(url2)
    soup1=BeautifulSoup(page1, 'html.parser')
    for div in soup1.find_all(class_="col-sm-10"):
        y=div.find('p')
        if y!= None :
            term_desc.append(y)
print(term_desc)
```

```python
term_desc2 = [] 
for val in term_desc: 
    if val != None : 
        term_desc2.append(val) 
```
